```python
import pandas as pd
import plotly.express as px
```

```python
# Lire le CSV
df = pd.read_csv("docs/_static/data.csv")

# Convertir la date
df["date"] = pd.to_datetime(df["date"])

# Créer le graphique
fig = px.line(
    df,
    x="date",
    y="valeur",
    title="Évolution des valeurs",
    markers=True
)

# Export HTML interactif
fig.write_html(
    "docs/_static/graph.html",
    include_plotlyjs="cdn",
    full_html=True
)

print("Graphique généré !")
```
